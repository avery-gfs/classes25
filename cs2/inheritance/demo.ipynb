```python
class Animal:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __repr__(self):
        return f"{self.name} is {self.age} years old"

    def have_birthday(self):
        self.age += 1
        print(f"{self.name} grew up and is now {self.age}")
```

```python
class Dog(Animal):
    def __init__(self, name, age, breed):
        self.breed = breed
        super().__init__(name, age)

    def __repr__(self):
        return f"{self.name} is a {self.breed} and is {self.age}"

    def bark(self):
        print(f"{self.name} is barking")
```

```python
class Bird(Animal):
    def __init__(self, name, age, color):
        self.color = color
        super().__init__(name, age)

    def eat(self):
        print(f"{self.name} is eating")
```

```python
woodpecker = Bird("bob", 3, "blue")
woodpecker.eat()
print(woodpecker)
```

```python
peanut = Dog("peanut", 2, "Bernadoodle")
print(peanut)
peanut.bark()
```

```python
for animal in [peanut, woodpecker]:
    animal.have_birthday()
    print(animal)
```

```python
print(isinstance(peanut, Dog))
print(isinstance(woodpecker, Dog))
print(isinstance(peanut, Animal))
```
