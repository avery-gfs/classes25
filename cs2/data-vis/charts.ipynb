```python
import polars as pl
import altair as alt

cities = pl.read_csv("data/cities.csv") # Load dataframe from CSV
```

```python
topCities = cities.head(20);

alt.Chart(topCities).mark_bar().encode(
  alt.X("pop2024"),
  alt.Y("city", sort = "-x"),
);
```

```python
cityCounts = cities.group_by("state").count();

alt.Chart(cityCounts).mark_bar().encode(
  alt.X("state", sort = "-y"),
  alt.Y("count"),
);
```

```python
pctChange = cities.with_columns(
  (pl.col("pop2024") / pl.col("pop2020") * 100 - 100).alias("pctChange"),
).sort("pctChange", descending = True).head(10);

alt.Chart(pctChange).mark_bar().encode(
  alt.X("pctChange"),
  alt.Y("city", sort = "-x"),
);
```

```python
top10 = cities.head(10);

alt.Chart(top10).mark_circle(size = 40).encode(
  alt.X("area"),
  alt.Y("pop2024"),
  alt.Color("city"),
);
```
